```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.patches as mpatches
```

```python
nivel1_df = pd.read_csv("nivel1python.csv")
nivel2_df = pd.read_csv("nivel2python.csv")
nivel3_df = pd.read_csv("nivel3python.csv")
```

```python
data = [
    nivel3_df['Pontuação final'], nivel2_df['Pontuação final'],
    nivel1_df['Pontuação final']
]

fig1, ax1 = plt.subplots(figsize=(12, 8), facecolor=(1, 1, 1))

flierprops = dict(markerfacecolor='c', marker='o', markersize=10)
meanlineprops = dict(linestyle='', linewidth=2, color='orange')
meanpointprops = dict(marker='D',
                      markeredgecolor='black',
                      markerfacecolor='firebrick',
                      markersize=10)
medianprops = dict(linestyle='-', linewidth=2, color='blue')

box = ax1.boxplot(data,
                  vert=False,
                  labels=['Nível 3', 'Nível 2', 'Nível 1'],
                  meanline=False,
                  showmeans=True,
                  widths=0.7,
                  patch_artist=True,
                  boxprops=dict(alpha=0.5),
                  flierprops=flierprops,
                  meanprops=meanpointprops,
                  medianprops=medianprops)

colors = ['lightgreen', 'orange', 'cyan']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

ax1.grid(b=True, which='major', linestyle=':', linewidth=1.0)
ax1.tick_params(which='both', labelsize=14)
ax1.set_xlabel('Notas', size=15)
ax1.set_xticks(np.arange(5, 101, 5))

fig1.subplots_adjust(top=0.90)
fig1.suptitle('Análise das notas - Comparação dos níveis', fontsize=20)

plt.show()

figname = 'images/boxplot_all'
fig1.savefig(figname, bbox_inches='tight', dpi=300)
```
